**Calculating error **

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
m = ols('mpg ~ horsepower + np.power(horsepower,2)',auto).fit()
print(m.summary())
```

 F statistics = 581(degree=1),401(degree=2) and the corresponding p-value is very small >> NULL HYPOTHESIS rejected

```python
# The R^2 is close to 0.7 so the relation betweeen predictor and response is strong
# It is a negetive relation
```

```python
m.predict(auto['horsepower']).head(5)
```

### Residual sum of squares

Let's calculate the residual sum of squares 

$$ S = \sum_{i=1}^N r_i = \sum_{i=1}^N (y_i - (\beta_0 + \beta_1 x_i))^2 $$

```python
RSS = np.sum((auto['mpg'] - m.predict(auto['horsepower'])) ** 2)
RSS
```

#### Calculating RSE

```python
p = 1
n = len(auto['mpg'])
RSE = np.sqrt(RSS/(n-p-1))
RSE
```
